```python
# Import necessary libraries
import pandas as pd
import numpy as np
```

```python
# Import adidas US retail CSV file
adidas_df = pd.read_csv('E:\\Data Analyst Mastery\\Adidas Products\\adidas_usa.csv')
adidas_df.head(10)
```

### CLEAN THE DATA 

```python
# Check the information on the dataframe
adidas_df.info()
```

```python
# Delete unneeded columns
del adidas_df['url']
del adidas_df['currency']
del adidas_df['source']
del adidas_df['source_website']
del adidas_df['breadcrumbs']
del adidas_df['brand']
del adidas_df['images']
del adidas_df['country']
del adidas_df['language']
del adidas_df['crawled_at']
adidas_df.head(10)
```

```python
# Rename each header column
adidas_df = adidas_df.rename({'name':'Name','sku':'SKU','selling_price':'Selling Price (USD)','original_price':'Original Price (USD)','availability':'Availability','color':'Color','category':'Category','description':'Description','average_rating':'Rating','reviews_count':'Reviews Count'}, axis=1)
adidas_df.head(10)
```

```python
# Find the strings '$' on Original Price (USD) column
adidas_df[adidas_df['Original Price (USD)'].str.contains('$')==True]
```

```python
# Remove the strings '$' on Original Price (USD) column
adidas_df['Original Price (USD)'] = adidas_df['Original Price (USD)'].str.replace('[$]', '', regex=True)
adidas_df.head(10)
```

```python
# Check column SKU if there is a duplicate
duplicateRows = adidas_df[adidas_df.duplicated(['SKU'])]
duplicateRows
```

```python
# Check each column if there is a null value
adidas_df.isnull().any()
```

```python
# Replace null value with 0
adidas_df['Original Price (USD)'] = adidas_df['Original Price (USD)'].fillna(0)
adidas_df.head(10)
```

```python
# Check the data type of each column
adidas_df.dtypes
```

```python
# Change the data type of each column as needed
convert = {'Original Price (USD)' : int, 'Selling Price (USD)' : int, 'Reviews Count' : int}
adidas_df = adidas_df.astype(convert)
adidas_df.head(10)
```

### EXPORT THE DATA 

```python
# Export DataFrame to Excel and CSV file
output_file_path_excel = 'E:\\Data Analyst Mastery\\Adidas Products\\adidasUSRetailCleaned.xlsx'
output_file_path_csv = 'E:\\Data Analyst Mastery\\Adidas Products\\adidasUSRetailCleaned.csv'
adidas_df.to_excel(output_file_path_excel, index=False)
adidas_df.to_csv(output_file_path_csv, sep='$', index=False)
```
